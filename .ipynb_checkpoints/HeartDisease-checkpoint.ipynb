```python
#Logistic Regression

y = heart.target.values

x_data = heart.drop(["target"],axis=1)
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data)).values
```

```python
x_data
```

```python
#Train and test the model
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.20,random_state=56)
```

```python
lr=LogisticRegression()
```

```python
#Train model
lr.fit(x_train,y_train)
print("test accuracy {}".format(lr.score(x_test,y_test)))
lr_score=lr.score(x_test,y_test)
```

```python
#Test model
y_prediction = lr.predict(x_test)
y_actual=y_test
cm = confusion_matrix(y_actual,y_prediction)
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel("Predictions Y")
plt.ylabel("Actual Y")
plt.show()
```

```python
#K Nearest Neighbors
knn=KNeighborsClassifier()
```

```python
#Train model, test accuracy same as logistic regression
knn.fit(x_train,y_train)
print("test accuracy {}".format(knn.score(x_test,y_test)))

knn_prediction_score=knn.score(x_test,y_test)
```

```python
#Test model
y_prediction = knn.predict(x_test)
y_actual=y_test
cm = confusion_matrix(y_actual,y_prediction)
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel("Predictions Y")
plt.ylabel("Actual Y")
plt.show()
#predictions are close to that of logistic regression
```
